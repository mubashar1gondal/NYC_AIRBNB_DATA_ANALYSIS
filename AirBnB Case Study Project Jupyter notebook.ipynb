```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.nunique().neighbourhood_group
```

```python
air_bnb['neighbourhood_group'].value_counts()
```

```python
# Are private rooms the most popular in manhattan?
air_bnb.groupby("neighbourhood_group").room_type.value_counts()
#entire home/apt are the most popular in NYC
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb.sort_values("reviews_per_month", ascending = False).head(10)
#row NYC seems to get the most revies per month
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby("neighbourhood_group").price.mean()
#manhattan has the highest average price
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby('neighbourhood_group').price.sum()
#manhattan also has the highest total price
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby('host_name').price.sum().sort_values(ascending = False).head()
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb.loc[(air_bnb['number_of_reviews'] >= 100) & (air_bnb['availability_365']==0)]

```

```python
# What host has the highest total of prices and where are they located?

air_bnb.groupby("host_id").price.sum().sort_values(ascending=False)
#host 2195177861
```

```python
# When did Danielle from Queens last receive a review?
air_bnb.loc[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens')].sort_values('last_review', ascending = False)
```
